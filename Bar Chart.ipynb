```python
# with csv data
import pandas as pd
from collections import Counter

data = pd.read_csv('data/data.csv')
```

```python
data.head()
```

```python
ids = data['Responder_id']
lang_responses = data['LanguagesWorkedWith']
```

```python
language_counter = Counter()
```

```python
language_counter
```

```python
for responses in lang_responses:
    language_counter.update(responses.split(';'))
```

```python
language_counter
```

```python
language_counter.most_common(15)
```

```python
languages = []
popularity = []

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])
```

```python


plt.barh(languages, popularity) # barh for horizontal chart

plt.title('Most Popular Languages')
plt.xlabel('Praogramming Languages')
plt.ylabel('Number of People Who Use')

plt.show()
```

```python
# To get reverse chart
languages.reverse()
popularity.reverse()

plt.barh(languages, popularity) # barh for horizontal chart

plt.title('Most Popular Languages')
plt.xlabel('Praogramming Languages')
plt.ylabel('Number of People Who Use')

plt.show()
```
