```python
from matplotlib import pyplot as plt

plt.style.use('fivethirtyeight')
```

```python
slices = [60, 40, 30, 20]
```

```python
plt.pie(slices)

plt.title('My Awesome Pie Chart')
plt.tight_layout()
plt.show()
```

```python
labels = ['Sixty', 'Fourty', 'Extra1', 'Extra2']
colors = ['Blue', 'red', 'yellow', 'green']
```

```python
plt.pie(slices, labels=labels, colors=colors, wedgeprops={'edgecolor': 'black'})

plt.title('My Awesome Pie Chart')
plt.tight_layout()
plt.show()
```

```python
# Pip Chart is good for comparing less than 5 features. else we Bar Chart is preferable
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0] # To explode Python. 0.1 means 10% radiopus exploded
```

```python
plt.pie(slices, labels=labels, explode=explode, wedgeprops={'edgecolor': 'black'})

plt.title('My Awesome Pie Chart')
plt.tight_layout()
plt.show()
```

```python
# to add shadow
plt.pie(slices, labels=labels, explode=explode, shadow=True, wedgeprops={'edgecolor': 'black'})

plt.title('My Awesome Pie Chart')
plt.tight_layout()
plt.show()
```

```python
# to rotate
plt.pie(slices, labels=labels, explode=explode, shadow=True, startangle=90, wedgeprops={'edgecolor': 'black'})

plt.title('My Awesome Pie Chart')
plt.tight_layout()
plt.show()
```

```python
# to display percentage
plt.pie(slices, labels=labels, explode=explode, 
        shadow=True, startangle=90, autopct='%1.1f%%',
        wedgeprops={'edgecolor': 'black'})

plt.title('My Awesome Pie Chart')
plt.tight_layout()
plt.show()
```
