```python
import pandas as pd
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates

plt.style.use('seaborn')
```

```python
dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]
```

```python
y = [0, 1, 3, 4, 6, 5, 7]
```

```python
dates
```

```python
plt.plot_date(dates, y)
plt.show()
```

```python
plt.plot_date(dates, y, linestyle='solid')
plt.show()
```

```python
plt.plot_date(dates, y, linestyle='solid')

# gcf = get current formate
plt.gcf().autofmt_xdate() # to formate a xlabel

plt.show()
```

```python
plt.plot_date(dates, y, linestyle='solid')

# gcf = get current formate
plt.gcf().autofmt_xdate() # to formate a xlabel

date_format = mpl_dates.DateFormatter('%b, %d %Y') # to fomate a date shown

# gca = get current axis
plt.gca().xaxis.set_major_formatter(date_format)

plt.show()
```

```python
data = pd.read_csv('data/time_series_data.csv')
```

```python
data.head()
```

```python
price_date = data['Date']
price_close = data['Close']
```

```python
plt.plot_date(price_date, price_close, linestyle='solid')

plt.gcf().autofmt_xdate() 

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()

plt.show()
```

```python
# to convert a date column to datetime method instead of string
data['Date'] = pd.to_datetime(data['Date'])

data.sort_values('Date', inplace=True) # sorting of data by date column

plt.plot_date(price_date, price_close, linestyle='solid')

plt.gcf().autofmt_xdate() 

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()

plt.show()
```
